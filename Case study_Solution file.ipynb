## 3. Explore the relationship between grip and arm scores and job performance measures (ratings and sim scores).

```python
import numpy as np
from scipy.stats import pearsonr

# Assuming your DataFrame is named df
# Extracting grip scores and job performance ratings
grip_scores = df['GRIP']
ratings = df['RATINGS']

# Performing Pearson correlation test
corr_coef, p_value = pearsonr(grip_scores, ratings)

# Output the correlation coefficient and p-value
print("Pearson correlation coefficient:", corr_coef)
print("P-value:", p_value)

# Set significance level(Not given in the question, hence an assumption)
alpha = 0.05

# Interpretation of the test
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant correlation between grip scores and job performance ratings.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between grip scores and job performance ratings.")
```

## 4. Use correlation analysis to quantify the strength and direction of associations between variables. Also draw the scatterplot. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named df
# Calculate correlation coefficients
correlation_matrix = df.corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Draw scatterplots
sns.pairplot(df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# Draw scatterplots
plt.figure(figsize=(10, 8))
sns.scatterplot(x='GRIP', y='ARM', data=df)
plt.title('Scatter Plot of Grip vs Arm Scores')
plt.xlabel('Grip Scores')
plt.ylabel('Arm Scores')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='GRIP', y='RATINGS', data=df)
plt.title('Scatter Plot of Grip vs Ratings')
plt.xlabel('Grip Scores')
plt.ylabel('Ratings')
plt.show()

plt.figure(figsize=(10, 8))
sns.scatterplot(x='GRIP', y='SIMS', data=df)
plt.title('Scatter Plot of Grip vs Sims Scores')
plt.xlabel('Grip Scores')
plt.ylabel('Sims Scores')
plt.show()

# Similarly, create scatter plots for other pairs of variables
```

Inference made:
> Strong Positive Correlation between ARM and GRIP (0.65):
* The strong positive correlation between ARM and GRIP scores (0.65) indicates a significant relationship between the strength of a person's arm and their grip strength.
* This suggests that individuals with higher ARM scores tend to have higher GRIP scores, and vice versa.
* This relationship implies that there may be common underlying factors or physiological traits influencing both arm and grip strength.

>Moderate Positive Correlation between SIMS-ARM and SIMS-GRIP (0.37, 0.34):
* The moderate positive correlation between SIMS-ARM (0.37) and SIMS-GRIP (0.34) indicates a moderate relationship between the scores obtained from simulations of physically demanding work tasks and both arm and grip strength.
* While not as strong as the correlation between ARM and GRIP, this suggests that performance in simulated physically demanding tasks is moderately associated with arm and grip strength.
* This implies that individuals with higher arm and grip strength may perform better in simulations of physically demanding work tasks.

>Weak Positive Correlation between Ratings and ARM, GRIP, and SIMS (0.18, 0.20, 0.14):
* The weak positive correlations between Ratings and ARM (0.18), GRIP (0.20), and SIMS (0.14) suggest a weak relationship between job performance ratings provided by supervisors and strength measures (arm, grip) and simulation scores.
* While statistically significant, these correlations indicate that job performance ratings have only a weak association with arm strength, grip strength, and performance in simulated tasks.
* This suggests that other factors beyond strength measures may play a more significant role in determining job performance ratings, such as skill level, technique, or non-physical attributes.
* Overall, the correlation analysis suggests that arm and grip strength are strongly related, while performance in simulated physically demanding tasks has a moderate association with strength measures. Job performance ratings, on the other hand, have only a weak association with strength measures, indicating that they may be influenced by other factors beyond physical strength.
